```python
def generate_race_chart(school_name, s):
    race_chart = pd.DataFrame()
    color_list = ['Beverages', 'Technology', 'Business Services', 'Diversified', 'Automotive', 'Media', 'Restaurants', 'Telecommunications', 'Tobacco']
    copy_of_cites_per_year = cites_per_year[cites_per_year['name'].isin(school_name)]
    copy_of_cites_per_year.drop([col for col, val in copy_of_cites_per_year.sum().iteritems() if val == 0], axis=1, inplace=True)
    x = list(copy_of_cites_per_year.columns)
    x.remove('name')
    for i in range(len(x)):
        x[i] = int(x[i])
    for i in range(min(x), 2021):
        for j in copy_of_cites_per_year.iterrows():
            new_l = {"date": str(i) + "-01-01", "name": j[1]['name'], "category": random.choice(color_list), "value": j[1][str(i)]}
            race_chart = race_chart.append([new_l])
    filename = "race_chart_" + s.strip("_")
    print(filename)
    race_chart.to_csv(filename, index = False)  
    return race_chart
```

```python
def generate_race_chart_total(school_name,s):
    race_chart_ = pd.DataFrame()
    color_list = ['Beverages', 'Technology', 'Business Services', 'Diversified', 'Automotive', 'Media', 'Restaurants', 'Telecommunications', 'Tobacco']
    copy_of_cites_per_year = cites_per_year[cites_per_year['name'].isin(school_name)]
    copy_of_cites_per_year.drop([col for col, val in copy_of_cites_per_year.sum().iteritems() if val == 0], axis=1, inplace=True)
    x = list(copy_of_cites_per_year.columns)
    x.remove('name')
    for i in range(len(x)):
        x[i] = int(x[i])
    for i in range(min(x), 2021):
        for j in copy_of_cites_per_year.iterrows():
            v = 0
            for kk in range(min(x), i + 1):
                v += j[1][str(kk)]
            new_l = {"date": str(i) + "-01-01", "name": j[1]['name'], "category": random.choice(color_list), "value": v}
            race_chart_ = race_chart_.append([new_l])
    filename = "race_chart_total_" + s.strip("_")
    print(filename)
    race_chart_.to_csv(filename, index = False)  
    return race_chart_
```

```python

SCHOOL_NAME = "NYUsh"
index = {}
repeated = []
nodes = []
links = []
k = 0
for i in n_names:
    if i not in index:
        if SCHOOL_NAME == "" or (SCHOOL_NAME == "NYUsh" and i in NYUsh_) or (SCHOOL_NAME == "NYUad" and i in NYUad_):
            index[i] = k
            the_size = total_citation[total_citation["name"] == i]["total_cite"][0] / 100
            if i in NYU_:
                nodes.append({"name": i, "group": 1, "size": the_size})
            elif i in NYUsh_:
                nodes.append({"name": i, "group": 2, "size": the_size})
            else:
                nodes.append({"name": i, "group": 3, "size": the_size})
            k += 1
print(index)
for i in connected_names.iterrows():
    c = i[1]["coauthorname"]
    n = i[1]["name"]
    if [c,n] not in repeated and n in index and c in index:
        repeated.append([c,n])
        links.append({"source": index[n], "target": index[c], "value": 3, "group": 1})
final = {}
final["nodes"] = nodes
final["links"] = links
print(final)
#final.to_csv("world_map.csv", index = False)
with open(SCHOOL_NAME + '.json', 'w') as fp:
    json.dump(final, fp)
```
